```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
list1 = [i for i in range(25)]
new_Index= [str(i) for i in list1]
data.columns = new_Index
data.head(5)
```

```python
for index in new_Index:
    data[index] = data[index].str.lower()
```

```python
data.head(5)
```

```python
headline=[]
for row in range(len(data.index)):
    headline.append(' '.join(str(x) for x in data.iloc[row,0:25])) 
```

```python
headline[3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headline)
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27])) 
```

```python
test_dataset = countvector.transform(test_transform)
test_dataset
```

```python
predictions = randomclassifier.predict(test_dataset)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
classificationReport = classification_report(test['Label'],predictions)
confusionMatrix = confusion_matrix(test['Label'],predictions)
accuracyScore = accuracy_score(test['Label'],predictions)
```
